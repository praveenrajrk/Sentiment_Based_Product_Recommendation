```python
# Write your function here to clean the text and remove all the unnecessary elements.
def text_cleaning(sentence):
  sentence = sentence.lower() #this will convert text to lowercase
  sentence = sentence.strip() # this will remove whitespaces in front and back of the sentence if any
  sentence = re.sub("\[\s*\w*\s*\]", "", sentence) # removing text that are present inside the square brackets if any
  table = "abc".maketrans('','', string.punctuation) # for removing all the punctuations 
  sentence = sentence.translate(table)
  sentence = re.sub("\S*\d\S*", "", sentence) #removing words containing numbers

  return sentence

```

```python
df_clean = product_reviews[['id','name', 'reviews_combined_text', 'user_sentiment']]
df_clean.head()
```

```python
df_clean["reviews_cleaned_text"] = df_clean.reviews_combined_text.apply(lambda x: text_cleaning(x))
```

```python
df_clean.head()
```

```python
## A mapper function to map NLTK parts_of_speech
def get_wordnet_pos(tag):
  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB
  elif tag.startswith('N'):
    return wordnet.NOUN
  elif tag.startswith('R'):
    return wordnet.ADV
  else:
    return wordnet.NOUN
```

```python
stop_words = set(stopwords.words('english'))
# function to remove stopwords
def fn_stopword_remove(sentence):
    words = [word for word in sentence.split() if word.isalpha() and word not in stop_words]
    return " ".join(words)
```

```python
lemmatizer_obj = WordNetLemmatizer()

# creating a function to lemmatize the words in the sentence
def fn_lemma_text(text):
  sentence_pos_tags = nltk.pos_tag(word_tokenize(fn_stopword_remove(text))) # first removing the stopwords, then tokenize the sentence using word_tokenize then extracting pos tags
  # By mapping pos tag and the corresponding word extracting lemma
  words =[lemmatizer_obj.lemmatize(tag[0], get_wordnet_pos(tag[1])) for idx, tag in enumerate(sentence_pos_tags)] 
  return " ".join(words)

```

```python
## extracting the lemmatized sentence
df_clean["reviews_lemmatized_text"] = df_clean.reviews_cleaned_text.apply(lambda sentence: fn_lemma_text(sentence))
```

```python
# creating a function to extract n-gram words and its frequency based on given range
def most_freq_ngrams(text, num = None, ngram = (1,1)):
    vector = CountVectorizer(stop_words='english', ngram_range= ngram).fit(text)
    bag_of_words = vector.transform(text)
    sum_of_words = bag_of_words.sum(axis=0)
    frequency_of_words = [(word,sum_of_words[0, index] )for word,index in vector.vocabulary_.items()]
    frequency_of_words = sorted(frequency_of_words, key = lambda x: x[1], reverse = True)
    return frequency_of_words[:num]
```

## Model Building Phase

```python
## Creating a general Model class to optimize code reusability 
class BuildModel:
  def __init__(self, model, x_train, x_test, y_train, y_test):
    self.model = model
    self.x_train = x_train
    self.x_test = x_test
    self.y_train = y_train
    self.y_test = y_test
  
  def train_and_predict_model(self):
    self.model.fit(self.x_train, self.y_train)
    return self.model.predict(self.x_test)

  def evaluate_model(self, y_pred_class):
    print("\n")
    print("#"*30)
    self.result_metrics = self.fetch_metrics(y_pred_class)
    print("#"*30)
    print("\n")
    
    self.print_classification_report(y_pred_class)
    print("#"*30)
    print("\n")
    self.print_confusion_matrix(y_pred_class)
        
    print("#"*30)
    metrics.plot_roc_curve(self.model, self.x_test, self.y_test) #for plotting roc curve
    
    return self.result_metrics

  def fetch_metrics(self, y_pred_class):
    result_metrics = [] 
    accuracy = metrics.accuracy_score(self.y_test, y_pred_class)
    precision = metrics.precision_score(self.y_test, y_pred_class)
    recall = metrics.recall_score(self.y_test, y_pred_class)
    f1score = metrics.f1_score(self.y_test, y_pred_class)
    y_pred_prob = self.model.predict_proba(self.x_test)[:,1]
    roc_auc = metrics.roc_auc_score(self.y_test, y_pred_prob)
    
    print(f"Accuracy is : {accuracy*100:.1f}%")
    print(f"Precision is : {precision*100:.1f}%")
    print(f"Recall is : {recall*100:.1f}%")
    print(f"F1 Score is : {f1score*100:.1f}%")
    print(f"Roc-Auc Score is:{roc_auc*100:.1f}%")
    
    result_metrics.append(accuracy)
    result_metrics.append(precision)
    result_metrics.append(recall)
    result_metrics.append(f1score)
    result_metrics.append(roc_auc)
    return result_metrics
  
  def print_classification_report(self, y_pred_class):
    print(metrics.classification_report(self.y_test, y_pred_class))

  def print_confusion_matrix(self, y_pred_class):
    print(metrics.confusion_matrix(self.y_test, y_pred_class))
               
    
  
```

Rating predicted by the user (for products rated as well as not rated) is the weighted sum of correlation with the product rating

## Top 20 Product Recommendations- Then filter based on Sentiment Model and give top 5 recommendations

```python
def get_sentiment_based_recommendations(user_name):
    if (user_name in user_final_rating.index):
        recommendations = list(user_final_rating.loc[user_name].sort_values(ascending=False)[0:20].index)#fetching top20 recommendations for the user
        temp = df_clean[df_clean.id.isin(recommendations)]
        #transfor the input data using tf-idf vectorizer that has been already built
        X =  tf_idf.transform(temp["reviews_lemmatized_text"].values.astype(str))
        temp["predicted_sentiment"]= log_reg_model.predict(X)
        temp = temp[['name','predicted_sentiment']]
        temp_grouped = temp.groupby('name', as_index=False).count()
        temp_grouped["pos_review_count"] = temp_grouped.name.apply(lambda x: temp[(temp.name==x) & (temp.predicted_sentiment==1)]["predicted_sentiment"].count())
        temp_grouped["total_review_count"] = temp_grouped['predicted_sentiment']
        temp_grouped['pos_sentiment_percent'] = np.round(temp_grouped["pos_review_count"]/temp_grouped["total_review_count"]*100,2)
        return temp_grouped.sort_values('pos_sentiment_percent', ascending=False)
    else:
        print(f"User name {user_name} doesn't exist")
```

```python
#test case that was already used -- user  name 
get_sentiment_based_recommendations("01impala")
```

```python
## extracting the top 5 recommendations based on positive sentiment percentage
get_sentiment_based_recommendations("01impala")[:5]
```

```python
## test the case where user name does not exist
get_sentiment_based_recommendations("test_user001")
```
